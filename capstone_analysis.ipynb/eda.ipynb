```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_house_prices.csv')

# 1. Price Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, color='blue')
plt.title('Property Prices ki Distribution')
plt.show()

# 2. Area vs Price (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Area', y='Price', hue='Location')
plt.title('Area vs Price (Location ke hisab se)')
plt.show()

# 3. Location-wise Average Price (Bar Chart)
plt.figure(figsize=(10, 6))
df.groupby('Location')['Price'].mean().sort_values().plot(kind='barh', color='teal')
plt.title('Har Location ka Average Price')
plt.show()

# 4. Property Type vs Price (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Property_Type', y='Price', data=df)
plt.title('Property Type ka Price par asar')
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Factors ke beech ka Correlation')
plt.show()
```
